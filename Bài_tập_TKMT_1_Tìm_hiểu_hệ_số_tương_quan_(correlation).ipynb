BÀI TẬP: Bạn hãy tìm hiểu về hệ số tương quan (correlation), cách đọc và giải thích ý nghĩa hệ số tương quan.


Tương quan (correlation) là chỉ số thống kê đo lường mức độ mạnh yếu của mối quan hệ giữa hai biến số.

Bản thân sự tương quan là một kỹ thuật toán học để kiểm tra mối quan hệ giữa hai biến định lượng, chẳng hạn như giá của một chiếc ô tô và kích thước động cơ của nó.

Giá trị dương biểu thị rằng cả hai biến chuyển động theo cùng một hướng trong khi giá trị âm cho chúng ta biết rằng chúng di chuyển theo các hướng ngược nhau.

Cách đọc

Hệ số tương quan có giá trị từ -1.0 đến 1.0. Kết quả được tính ra lớn hơn 1.0 hoặc nhỏ hơn -1 có nghĩa là có lỗi trong phép đo tương quan.

- Hệ số tương quan có giá trị âm (biểu đồ bên trái) cho thấy hai biến có mối quan hệ nghịch biến hoặc tương quan âm (nghịch biến tuyệt đối khi giá trị bằng -1)
- Hệ số tương quan có giá trị dương (biểu đồ bên phải) cho thấy mối quan hệ đồng biến hoặc tương quan dương (đồng biến tuyệt đối khi giá trị bằng 1)
- Tương quan bằng 0 cho hai biến độc lập với nhau (biểu đồ ở giữa). Lấy một ví dụ cụ thể về hai biến x và y, đó chính là biến x (Mức độ hài lòng), biến y (Tiền lương của nhân viên).

Tương quan được kết nối chặt chẽ với các đại lượng thống kê khác như giá trị trung bình, độ lệch chuẩn, phương sai và hiệp phương sai

Có các hệ số tương quan: 
- Pearson
- Spearman
- Kendall

HỆ SỐ TƯƠNG QUAN PEARSON

- Có thể được sử dụng để tóm tắt độ mạnh của mối quan hệ tuyến tính giữa hai mẫu dữ liệu.

- Hệ số tương quan Pearson được tính bằng hiệp phương sai của hai biến chia cho tích của độ lệch chuẩn của mỗi mẫu dữ liệu


Hệ số tương quan Pearson (r) có giá trị giao động trong khoảng liên tục từ -1 đến +1:

- r = 0: Hai biến không có tương quan tuyến tính
- r = 1; r = -1: Hai biến có mối tương quan tuyến tính tuyệt đối.
- r < 0: Hệ số tương quan âm. Nghĩa là giá trị biến x tăng thì giá trị biến y giảm và ngược lại, giá trị biến y tăng thì giá trị biến x giảm.
- r > 0: Hệ số tương quan dương. Nghĩa là giá trị biến x tăng thì giá trị biến y tăng và ngược lại, giá trị biến y tăng thì giá trị biến x cũng tăng

HỆ SỐ TƯƠNG QUAN SPEARMAN

- Có thể được sử dụng để tóm tắt độ mạnh của mối quan hệ tuyến tính giữa hai mẫu dữ liệu.
- được sử dụng khi hai biến x và y không tuân theo luật phân phối chuẩn, trái ngược với hệ số tương quan Pearson
- Hệ số này được ước tính bằng cách biến đổi biến x, y thành biến có thứ bậc (rank), sau đó xem xét độ tương quan giữa hai dãy số có bậc này
- Tương quan hạng Spearman giữa x và y nói đơn giản là xem xét tính đơn điệu của 2 biến này với nhau. Nếu hệ số tương quan dương thì kết luận là x tăng y cũng tăng. Nếu hệ số tương quan âm thì kết luận là x tăng y giảm

HỆ SỐ TƯƠNG QUAN KENDALL

- Là một phương pháp phân tích phi tham số được ước tính bằng cách tìm các cặp số (x,y) “song hành” (concordant) với nhau. Một cặp (x,y) song hành ở đây được định nghĩa là có hiệu số (độ khác biệt) trên trục hoành có cùng dấu hiệu (dương hay âm) với hiệu trên trục tung. Nếu hai biến số x, y không có liên hệ với nhau thì số cặp song hành bằng hay tương đương với số cặp không song hành
- Có thể được sử dụng để tóm tắt độ mạnh của mối quan hệ tuyến tính giữa hai mẫu dữ liệu.
- hệ số này được sử dụng tương tự như spearman, thông thường hệ số này nhỏ hơn spearman
- Hệ số kendall ít dùng hơn so với 2 hệ số tương quan trên
