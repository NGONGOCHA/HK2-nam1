```python
import torch
a = torch.Tensor(np.array([1, 7]))  #Tensor hạng 1
print(a)
print(a.ndim)
```

```python
b = torch.Tensor(np.array([[1, 4],[2,3]]))    #Tensor hạng 2
print(b)
print(b.ndim)
```

```python
c = torch.Tensor(np.random.randint(0, 1000, size = 280)).reshape(5, 7, 8)   #Tensor hạng 3
print(c)
print(c.ndim)
```

```python
d = torch.Tensor(np.random.randint(0, 1000, size = 1120)).reshape(5, 7, 8, 4)   #Tensor hạng 4
print(d)
print(d.ndim)
```
